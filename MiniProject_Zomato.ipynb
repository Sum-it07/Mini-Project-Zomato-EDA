# **1.Merge the country sheet and Zomato file to get the country code**

```python
merged=pd.merge(df,c_code,on='Country Code')
```

```python
merged.head()
```

# **2) Get the Top rated restaurants in each city in India(On the basis of rating and votes).**

```python
top_rated=merged.loc[merged['Country']=='India']
top_rated_rest=top_rated.sort_values(by=['Aggregate rating','Votes'],ascending=False)
top_rated_rest_city=top_rated_rest.groupby('City').first()
top_rated_rest_city
```

# **3) Finding out relationship between the rating and votes.**

```python
sns.scatterplot(x='Aggregate rating',y='Votes',data=top_rated_rest_city)
```

```python
sns.scatterplot(x='Aggregate rating',y='Votes',data=merged)
```

# **6) cheap but best restaurants available in a city**

```python
merged.columns
```

```python
merged[['Country','City','Average Cost for two','Currency','Price range','Aggregate rating','Votes']]
```

```python
cities=pd.DataFrame(set(merged['City']))
for i in range(len(cities)):
  city_restaurants = merged[merged['City'] == cities.iloc[i,0]]
  cheap_best_restaurants = city_restaurants.sort_values(by=['Aggregate rating', 'Average Cost for two'], ascending=[False, True])
  print(cheap_best_restaurants[['Restaurant Name', 'Aggregate rating', 'Average Cost for two', 'Currency', 'Price range']].head(1))
```

##8) Aggregate rating of all the restaurants in each city in a country.

```python
average_rating_by_city_country = merged.groupby(['Country', 'City'])['Aggregate rating'].mean().reset_index()
display(average_rating_by_city_country)
```

##9) Does rating influences the cost of restaurant? (Draw boxplot for this).


```python

rounded_rating = merged['Aggregate rating'].round()
filtered_data = merged[rounded_rating.isin([1, 2, 3, 4, 5])]
plt.figure(figsize=(12, 6))
sns.boxplot(x=rounded_rating[rounded_rating.isin([1, 2, 3, 4, 5])],
            y=filtered_data['Average Cost for two'])
plt.title('Relationship between Aggregate Rating and Average Cost for Two')
plt.xlabel('Aggregate Rating (Rounded)')
plt.ylabel('Average Cost for Two')
```

Conclusion: The cost distribution looks quite similar across rating levels

##10) Top Percentage cover of restaurants in a city.(Draw pie graph for this).

```python
restaurant_counts = merged['City'].value_counts()
city_percentages = restaurant_counts / restaurant_counts.sum() * 100

top_n = 5
top_cities = city_percentages.head(top_n)
other_cities_percentage = city_percentages.iloc[top_n:].sum()

if other_cities_percentage > 0:
    top_cities['Other'] = other_cities_percentage

plt.figure(figsize=(10, 10))
plt.pie(top_cities, labels=top_cities.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Percentage Cover of Restaurants by City')
plt.axis('equal')
plt.show()
```

##11) Top cuisines in Indian restaurants.( Draw pie graph for this)

```python
indian_restaurants = merged[merged['Country'] == 'India']
cuisine_counts = indian_restaurants['Cuisines'].value_counts()

top_n = 10
top_cuisines = cuisine_counts.head(top_n)
other_cuisines_percentage = cuisine_counts.iloc[top_n:].sum()

if other_cuisines_percentage > 0:
    top_cuisines['Other'] = other_cuisines_percentage

plt.figure(figsize=(10, 10))
plt.pie(top_cuisines, labels=top_cuisines.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Cuisines in Indian Restaurants')
```
